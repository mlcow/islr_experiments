# $\sigma(x)$ 

The journey to sigmoid.
- $x$ being linear
- $e^x$ is exponential, but has a min of $0$
- $1 + e^x$ isn't much off from $e^x$
- $\frac{1}{1+e^x}$ doesn't violate min of $0$, and also brings in a max of $1$
- $\sigma(x)$ is but just the mirror ($f(-x)$) of the previous func.

### $e^x$

```
#@markdown # What up!
start = -100 #@param {type:"slider", min:-100, max:100, step:1}
end = 100 #@param {type:"slider", min:-100, max:100, step:1}
step = 0.201 #@param {type:"slider", min:0.001, max:10, step:0.1}


x = np.arange(start, end, step)
plt.plot(x, np.exp(x), '-+')
```

### Inverse of add $1$ to $e^x$
$$\frac{1}{1 + e^x} = \frac{e^{-x}}{1+e^{-x}}$$

Hence the below plot looks like the inverse of the sigmoid!

```
plt.plot(x, 1/(1+np.exp(x)))
```

### The sigmoid 
$$\frac{e^x}{1 + e^x}$$

```
plt.plot(x, np.exp(x) / (1 + np.exp(x)))
```

### Both in a single plot

```
plt.plot(x, 1 / (1 + np.exp(x)))
plt.plot(x, np.exp(x) / (1 + np.exp(x)))
```

```
#@markdown # Altair plot
import pandas as pd
import altair as alt

# Simple line plot
# https://altair-viz.github.io/gallery/simple_line_chart.html
# Multiline plot
# https://github.com/altair-viz/altair/issues/968
```

```
data = pd.DataFrame({'x': x,
                     'y1': 1 / (1 + np.exp(x)),
                     'y2': np.exp(x)  / (1 + np.exp(x)) })

base = alt.Chart(data).mark_line().encode(x='x')

alt.layer(
    base.mark_line(color='blue').encode(y='y1'),
    base.mark_line(color='green').encode(y='y2')
).interactive(bind_y=False, bind_x=True)
```
